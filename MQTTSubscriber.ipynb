```python
import paho.mqtt.client as mqtt
import json
from datetime import datetime

BROKER = "broker.hivemq.com"

def on_message(client, userdata, msg):
    data = json.loads(msg.payload.decode())
    ts   = datetime.now().strftime("%H:%M:%S")

    if msg.topic == "bridge/sensor/alert":
        print(f"\n🚨 [{ts}] ALERT — Severity: {data.get('severity')} | Score: {data.get('score')}")
        print(f"   {data.get('message')}\n")
    else:
        print(f"[{ts}] ax={data['ax']:>7}  ay={data['ay']:>7}  az={data['az']:>7}  temp={data['temp_c']}°C")

sub = mqtt.Client(client_id="colab-dashboard")
sub.on_message = on_message
sub.connect(BROKER, 1883, 60)
sub.subscribe("bridge/sensor/data")
sub.subscribe("bridge/sensor/alert")

print("Dashboard listening...\n")
sub.loop_forever()
```
